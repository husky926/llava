```python
import json

def filter_cxr_items(data):
    return [item for item in data if "CXR" in item.get("image", "")]

# Load JSON file
def load_json(file_path):
    with open(file_path, "r") as file:
        return json.load(file)

# Save JSON file
def save_json(data, file_path):
    with open(file_path, "w") as file:
        json.dump(data, file, indent=4)

# Sample JSON input file path
input_file = "/content/MMed-RAG/data/training/alignment/radiology/radiology_report.json"
output_file = "/content/MMed-RAG/data/training/alignment/radiology/radiology_report_new.json"

# Process JSON
data = load_json(input_file)
filtered_data = filter_cxr_items(data)
save_json(filtered_data, output_file)

# Output filtered JSON
print(json.dumps(filtered_data, indent=4))
```
